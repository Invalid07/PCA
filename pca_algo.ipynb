```python
x=df
```

```python
y=dataset.target
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_scaler=s.fit_transform(x)
pd.DataFrame(x_scaler)
```

## step2 covariance matrix

<!-- The covariance matrix is a square matrix that summarizes the variances and covariances of different features 
in a dataset.The covariance matrix is a symmetric matrix that summarizes the pairwise covariances between different  features in a dataset. -->
   

##'''The covariance matrix is a square matrix that summarizes the variances and covariances of different features 
in a dataset.The covariance matrix is a symmetric matrix that summarizes the pairwise covariances between different 
features in a dataset.'''

```python
cov_mat=np.cov(x_scaler.T)
pd.DataFrame(cov_mat)
```

```python
eig_val,eig_vec =np.linalg.eig(cov_mat)
pd.DataFrame(eig_val)
```

```python
pd.DataFrame(eig_vec)
```

## variance 

```python
tot= sum(eig_val)
var_exp=[(i/tot)*100 for i in sorted(eig_val,reverse=True)]
```

```python
var_exp
```

##### cumulative variance

```python
cum_var=np.cumsum(var_exp)
pd.DataFrame(cum_var)
```

```python
plt.figure(figsize=(10,7))
plt.step(range(len(cum_var)),cum_var,label="cumu_expl_var")
plt.bar(range(len(var_exp)),var_exp,label="ind_expl_vari",color="g")
plt.xlabel("pc")
plt.ylabel("variance")
plt.legend()
plt.show()



```

```python
plt.bar(range(len(var_exp)), var_exp, label='individual explained variance', color = 'g')
```

```python
plt.figure(figsize=(10, 5))
plt.step(range(len(cum_var)), cum_var,label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend()
plt.show()
```
